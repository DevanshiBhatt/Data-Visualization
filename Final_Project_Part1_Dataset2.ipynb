### Dataset2 - Biopics

This dataset is taken from https://github.com/fivethirtyeight/data. The source of this data iss imdb. As mentioned in their GitHub repository, the dataset is available under the Creative Commons Attribution 4.0 International License. 

The Biopics dataset is a small summary of biopics released from 1915 to 2014. A biopic is basically a biographical movie. It is a non-fictional movie based on a historical person’s real life. The dataset is small in size with 14 columns and 761 rows. It contains interesting information about the name of subject, title of the movie which is based on the subject, subject’s race and sex, number of subjects, director of the movie, lead actor/actresses and the box office collection of the movie.  

```python
# Import the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("F:/DEVANSHI/UIUC/sem2/IS590 Data Viz/finalProject/biopics.csv", 
                 encoding='ISO-8859-1', 
                 na_values={'box_office':'-', 'type_of_subject':'Other'})
# data = data.fillna(0)
data.shape
data
```

```python
#data.drop(data["type_of_subject"]=='Other', inplace=True)
# data.size
```

```python
by_subject = data.groupby("type_of_subject")["type_of_subject"].size().nlargest(10, keep="first")
by_subject
```

```python
plt.rcParams["figure.dpi"] = 100              #changing the resolution of img 
plt.rcParams["figure.figsize"] = [9, 4.8]   #changing the figure size
```

```python
plt.bar(by_subject.index.values,by_subject.values, color='red')
plt.xlabel("Subject Type")
plt.ylabel("Number of Biopics")
#plt.xticks(np.arange(0,20,1))
#plt.yscale("log")
plt.title("Top 10 categories in Biopics")
```

```python
# by_race = pd.DataFrame(data.groupby(["subject_race", "subject_sex"]).size()).reset_index
by_race = data.groupby(["subject_race", "subject_sex"]).size()
by_race = by_race.unstack()
#by_race.to_frame().reset_index()
```

```python
by_race.plot.bar(stacked=True)
#plt.bar(data["subject_race" ])
```

```python
data.groupby("country")["number_of_subjects"].sum()
```

```python
by_year = data.groupby("year_release")["number_of_subjects"].sum()
by_year
```

```python
plt.plot(by_year, marker='o')
plt.xlabel("Year of Release")
plt.ylabel("Number of Biopics")
plt.xticks(np.arange(1915,2014,10))
#plt.yscale("log")
plt.title("Biopics by year")
```
